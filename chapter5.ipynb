```python
from sklearn.datasets import load_iris
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC, LinearSVR
from sklearn.svm import SVC, SVR
import numpy as np
```

```python
iris = load_iris(as_frame=True)
X = iris.data[["petal length (cm)", "petal width (cm)"]].values
y = (iris.target == 2)

svm_clf = make_pipeline(StandardScaler(), LinearSVC(C=1, random_state=42))
svm_clf.fit(X, y)
```

```python
X_new = [[5.5, 1.7],[5.0,1.5]]
svm_clf.predict(X_new)
```

```python
svm_clf.decision_function(X_new)
```

### Nonlinear SVM Classification

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures

X, y = make_moons(n_samples=100, noise=0.15, random_state=42)

polynomial_svm_clf = make_pipeline(PolynomialFeatures(degree=3), StandardScaler(), LinearSVC(C=10, max_iter=10000, random_state=42))

polynomial_svm_clf.fit(X, y)
```

### Polynomial Kernel

```python
from sklearn.svm import SVC

poly_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, coef0=1, C=5))
# the coef0 parameter tells us how much we are willing to let the high-degree terms have a bigger impact compared to the low-degree terms
poly_kernel_svm_clf.fit(X, y)
```

#### Gaussian RBF Kernel

```python
rbf_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", gamma=5, C=0.001))
## increasing gamma makes the bell-shaped curve narrower
rbf_kernel_svm_clf.fit(X, y)
```

### SVM Regression

```python
np.random.seed(42)
m = 50
X = 2 * np.random.rand(m, 1)
y = (4 + 3 * X + np.random.randn(m, 1)).ravel()
```

```python
svc_reg = make_pipeline(StandardScaler(), LinearSVR(epsilon=0.5, random_state=42))
svc_reg.fit(X,y)
```

```python
np.random.seed(42)
m = 100
X = 2 * np.random.rand(m, 1) - 1
y = (0.2 + 0.1 * X + 0.5 * X**2 + np.random.randn(m, 1)/10).ravel()
```

```python
svm_poly_reg = make_pipeline(StandardScaler(),
                             SVR(kernel="poly", degree=2, C=0.01, epsilon=0.1))
svm_poly_reg.fit(X, y)
```
